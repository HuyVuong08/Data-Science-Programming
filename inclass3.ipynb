```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

tips =sns.load_dataset('tips')
tips.head()
```

```python
tips['total_bill']
```

```python
plt.plot(tips['total_bill'], tips['tip'])
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

ax1.plot(tips['total_bill'], tips['tip'])
ax2.scatter(tips['total_bill'], tips['tip'])

ax1.set_title("Lines")
ax2.set_title("Scatter")

ax1.set_ylabel("Tips")
ax2.set_ylabel("Tips")

ax1.set_xlabel("Total_bill")
ax2.set_xlabel("Total_bill")
```

```python
plt.hist(tips["total_bill"], bins=10, density=False)
```

```python
plt.boxplot([tips[tips['sex'] == "Female"]["tip"], tips[tips['sex'] == "Male"]["tip"]])

plt.xlabel("Tip")
plt.ylabel("Sex")
```

```python
boxplot = plt.figure()

axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex'] == "Female"]["tip"], tips[tips['sex'] == "Male"]["tip"]], tick_labels=["Female", "Male"])

axes1.set_xlabel("Sex")
axes1.set_ylabel("Tips")
```

```python
def recode_sex(sex):
    if sex == "Female":
        return 2
    else:
        return 5

tips["sex_color"] = tips["sex"].apply(recode_sex)
tips

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(x=tips["total_bill"], y=tips["tip"], s=tips["size"]*20, c=tips["sex_color"], alpha=0.5)
```

```python
ax = sns.histplot(tips["total_bill"], kde=True)

ax.set_title("Total Bill Histogram with Density Plot")
plt.show()
```

```python
fg = sns.displot(data=tips, x="total_bill", col="sex", kde=True)

fg.axes.set_title("Total thing")
fg.axes.set_xlabel("Total X")
```

```python
ax = sns.countplot(data=tips, x="day")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", markers=["o", "o"])
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="day", row="smoker", hue="sex", markers=["o", "o"])
```

```python
j = sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")
j.set_axis_labels(xlabel="Total Bill", ylabel="Tip")

j.figure.suptitle("Joint Plot of Total Bill and Tip by Sex", fontsize=10, y=1.)
```

```python
sns.barplot(data=tips, x="time", y="total_bill")
```

```python
sns.violinplot(data=tips, x="time", y="total_bill", hue="sex", split=True)
```
