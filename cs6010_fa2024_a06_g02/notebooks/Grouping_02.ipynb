# Occupation


### Step 1. Import the necessary libraries


```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).


### Step 3. Assign it to a variable called users.


```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

users = pd.read_csv(url, delimiter='|')

print(users.head())
```

### Step 4. Discover what is the mean age per occupation


```python
mean_age_per_occupation = users.groupby('occupation')['age'].mean().round(2)

print(mean_age_per_occupation)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least


```python
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)

gender_counts['Male Ratio'] = gender_counts['M'] / (gender_counts['M'] + gender_counts['F'])

# Sort by Male Ratio from most to least
sorted_male_ratio = gender_counts.sort_values(by='Male Ratio', ascending=False)

sorted_male_ratio[['Male Ratio']]
```

### Step 6. For each occupation, calculate the minimum and maximum ages


```python
age_stats = users.groupby('occupation')['age'].agg(['min', 'max']).reset_index()

age_stats
```

### Step 7. For each combination of occupation and gender, calculate the mean age


```python
mean_age = users.groupby(['occupation', 'gender'])['age'].mean().reset_index()

# Rename the column for clarity
mean_age.rename(columns={'age': 'mean age'}, inplace=True)

mean_age
```

### Step 8. For each occupation present the percentage of women and men


```python
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)

gender_counts['total'] = gender_counts.sum(axis=1)

gender_counts['male percentage'] = (gender_counts['M'] / gender_counts['total']) * 100
gender_counts['female percentage'] = (gender_counts['F'] / gender_counts['total']) * 100

# Round the percentages to 2 decimal places
gender_counts['male percentage'] = gender_counts['male percentage'].round(2)
gender_counts['female percentage'] = gender_counts['female percentage'].round(2)
 
percentage_df = gender_counts[['male percentage', 'female percentage']].reset_index()

print(percentage_df)
```
